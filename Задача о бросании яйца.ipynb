# Условие задачи

Есть 2 яйца одинаковые по прочности. Требуется за минимальное число попыток
определить минимальную высоту дома (в этажах), сбрасывая с которой, яйцо разбивается.
Известно, что яйца точно разобьются, если их сбросить со 101-этажного дома, и не
разобьются, если их спросить с 0-этажного дома (с земли).
Придумайте оптимальный алгоритм с точки зрения количества попыток. Сколько при этом
потребуется попыток в худшем случае?

## Решение

Так как число яиц строго ограничено (2), решить задачу оптимальным бинарным поиском не получится.
Тогда рассмотрим следующую стратегию:
1. Допустим, мы бросаем первое яйцо с любым другим шагом n, начиная снизу. Тогда после того, как яйцо разобьтся, нам в худшем случае будет достаточно n-1 броска, чтобы определить нужный этаж.

![Иллюстрация к стратегии 1](strategy_1.png)

2. Что если мы зафиксируем максимальное количество бросков следующим образом: 

Мы бросили яйцо в первый раз на этаже 15, и оно разбилось, тогда нам осталось сделать максимум 14 бросков, чтобы найти нужный этаж. Если яйцо не разбилось, то мы делаем второй бросок через 14 этажей, и, если оно разбилось, то для поиска нужного этажа нам нужно будет совершить максимум 13 бросков. В обоих случаях количество бросков будет равно 15.
Таким образом, при выборе правильного начального шага, максимальное количество бросков будет одинпковым, где бы ни разбилось первое яйцо.
